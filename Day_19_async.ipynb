```python
# Task 1: Print the weather in cape town
# Create a function that takes in the city name and provide the temperature

def get_temperature(city_name):
    link = f'http://api.weatherapi.com/v1/current.json?key=fb838521799749b48ad93208241503&q={city_name}&aqi=no'
    response = requests.get(link, verify=False)
    weather = response.json()
    return weather['location']['name'], weather['current']['temp_c']
```

```python
get_temperature('cape town')
```

```python
# Task 2: Write a function that takes in a list of cities, and return the corresponding temperatures in a dictionary

def get_temperatures(cities):
    return {city: get_temperature(city)[1] for city in cities}

cities = ["Cape town", "Johannesburg", "Durban", "Pretoria", "Stellenbosch", "Somerset", "Khayelitsha" ] * 10
get_temperatures(cities)
```

## Slow response
- The server handles one request at a time -> sync
- sends requests, python waits for the server to respond before it can continue 
- synchronous requests will bottleneck and queue up
- By default everything is synchronous in python unless you specify


## Synchronise code
- Do things in order -> step by step.


## Asynchronise Code
- Do things in simalteneously
- while waiting for the server to respond, I can do another task at the mean time


## Single thread
- A thread is a small set of instructions that runs independently in the main process
- In each thread I can run one process -> when you run something


## Multi-thread
- Multithreading is a CPU feature that enables the execution of multiple threads simultaneously using the same resources.
- very difficult to program 

## Deadlock
- when two tasks want to access the same variable simultaneously.


## What we want
- We want single thread with asynchronous
