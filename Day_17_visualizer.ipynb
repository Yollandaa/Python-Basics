```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
tips_df = pd.read_csv('tips.csv')
tips_df
```

```python
# Add extra column tip_percentage
tips_df['tip_percentage'] = ((tips_df['tip'] / tips_df['total_bill']) * 100).round(2)
```

## Bar Graph

```python
x_days = tips_df['day'].unique()  #what you want for the x-axis
y_avg_total_bill = tips_df.groupby('day')['total_bill'].mean()  # what to plot for the y-axis
# x_days = y_avg_total_bill.index # same results for x-axis

plt.bar(x_days, y_avg_total_bill, color='crimson', alpha= 0.5)   #Draws the graph, you can change the color
plt.xlabel("Days")  # label the x-axis
plt.ylabel("average sales in R")  #labe the y axis
plt.title('Average sales in the week')
```

```python
# Plot using only one line
y_avg_total_bill.plot(kind = 'bar', xlabel ="Days", ylabel="Average sales in R", title="Average saes in a week", color='crimson')
```

## Pie Charts
- To get the percentage  -> autopct='%1.1f%%'

```python
y_avg_total_bill.plot(kind = 'pie',autopct='%3.2f%%', xlabel ="Days", ylabel="Average sales in R", title="Average sales in a week", color='crimson')
```

```python
# Task
# Create a pie chart comparing how many times people have visited on what time of the day
time_visited = tips_df['time'].value_counts()

time_visited.plot(kind = 'pie',autopct='%1.1f%%', xlabel ="Proportion of Meals", color='crimson')
```

### Scatterplots
- Shows outliers
- Shows trends
- Shows how the data is distributed

```python
#Tip percentage vs total_bill
plt.scatter(tips_df['total_bill'], tips_df['tip_percentage'])
plt.xlabel("Tip percentage %")  # label the x-axis
plt.ylabel("Total bill (R)")  #labe the y axis
plt.xticks(rotation=45) #rotate the labels
```

## Histogram
- Shows the spending patterns.
- How the total_bill is distributed
- More frequently people are spending $18 on the total bill.

```python
# How big your order size: ticket_size - total_bill -> x-axis | frequency -> y-axis
plt.hist(tips_df['total_bill'], color ='orange', bins=20)  #bins -> how much to divide
plt.xlabel('Total bill (R)')
plt.ylabel('Frequency')
```

## Seaborn
- The bar is average
- The stick is the maximum and minimum

```python
sns.barplot(x ='day', y ='total_bill', data=tips_df, palette='pastel')
plt.xlabel('Average sales (R)')
plt.ylabel('days')
plt.title("")
```

## Violin plot

```python
sns.violinplot(x = 'total_bill', y ='tip', data= tips_df,color='green')
plt.xlabel('Tip (R)')
plt.ylabel('Total bill (R)')
plt.title("Total bill vs. Tip with Regression Line")
```
